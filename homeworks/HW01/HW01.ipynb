##**1** Создание одномерного массива

```python
import numpy as np
array_1_1 = np.array([100, 200, 300, 400, 500, 600, 700, 800, 900, 1000])
array_1_2 = np.array([50, 100, 150, 200, 250, 300, 350, 400, 450, 500])
print(array_1_1)
print(array_1_2)
```

##**2** Двумерный массив, заполненный нулями

```python
import numpy as np
array_2 = np.zeros((3, 3))
print(array_2)
```

##**3** Двумерный массив, заполненный единицами

```python
import numpy as np
array_3 = np.ones((3, 3))
print(array_3)
```

##**4** Одномерный массив,(с помощью arange)

```python
import numpy as np
array_4 = np.arange(1,25)
print(array_4)
```

##**5** Форма данных для массива array_3

```python
print("Форма массива:", array_3.shape)
```

##**6**Тип данных для массива array_3

```python
print("Тип данных:", array_3.dtype)
```

##**7**Вывод первого и второго элемента массива array_4

```python
first_element = array_4[0]
second_element = array_4[1]
print(first_element)
print(second_element)
```

##**8** Нахождение 5-ого элемента массива array_4 (индексирование)

```python
index_array = array_4[4]
print(index_array)
```

##**9** Срез с первого по 23(включительно) элемент с шагом 3

```python
slice_of_array = array_4[0:23:3]
print(slice_of_array)
```

##**10** Поэлементное суммирование массивов

```python
sum_of_array = array_1_1 + array_1_2
print(sum_of_array)
```

##**11** Поэлементное умножение массивов

```python
multiplication_of_array = array_1_1 * array_1_2
print(multiplication_of_array)
```

##**12** Нахождение среднего значения массива

```python
mean_value = np.mean(array_1_1)
print(mean_value)
```

##**13** Нахождение медианного значения массива

```python
median_value = np.median(array_1_1)
np.median(array_1_1)
print(median_value)
```
